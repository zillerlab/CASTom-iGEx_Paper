```r
library(data.table)

setwd('/psycl/g/mpsziller/lucia/')

fold <- 'priler_project/Figure_CASTom-iGEx_paper/PriLer/'

summary_priler <- fread(sprintf('%sPriLer_summary_tissueSpecific.txt', fold), h=T, stringsAsFactors = F, data.table = F)
KS_DHS_compare <- fread(sprintf('%sKS_test_distribution_notzero.txt', fold), h=T, stringsAsFactors = F, data.table = F)
DHS_intersection <- fread(sprintf('%spercentage_DNAase_intersection_tissueSpec.txt', fold), h=T, stringsAsFactors = F, data.table = F)

DHS_intersection <- DHS_intersection[DHS_intersection$type == 'PriLer (reliable Genes)',]
DHS_intersection$tissue[DHS_intersection$tissue == 'Cells_EBV.transformed_lymphocytes']  <- 'Cells_EBV-transformed_lymphocytes'
```
